# Useful relations, how to get a column y from a column x

```python
print("-------------------------")
print("Even numbers")
print("X","\t","Y")
print("-------------------------")
for x in range(1,11):
    y = 2*x
    print(x,"\t",y)

print("-------------------------")
print("Odd numbers")
print("X","\t","Y")
print("-------------------------")
for x in range(1,11):
    y = 2*x-1
    print(x,"\t",y)

print("-------------------------")
print("Reverse Order")
print("X","\t","Y")
print("-------------------------")
for x in range(1,11):
    y = 11-x 
    print(x,"\t",y)

print("-------------------------")
print("Sum")
print("X","\t","Y")
print("-------------------------")
for x in range(1,11):
    y = int((x*(x+1))/2)  
    print(x,"\t",y)

print("-------------------------")
print("Grouped numbers")
print("X","\t","Y")
print("-------------------------")
n = 3 # number of group's elements 
for x in range(1,11):
    y1 =  int( (x-1)/(n))  * (n) +1 ;
    y2 =  y1 + n-1
    print(x,"\t",y1,":",y2)


print("-------------------------")
print("Fibonacci numbers")
print("X","\t","Y")
print("-------------------------")
# can we do it without a loop?
def fib(x):
  if (x == 1 or x ==2):
     f = 1      
  else:
     a = 1 ; b =1;
     for j in range(x-2):
         f = a + b
         a = b; b = f
  return f

for x in range(1,11):
    y = fib(x)
    print(x,"\t",y)

```
